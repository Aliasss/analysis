### 선호 가격대 전처리

구매 의사 없음 : 0  
1만원 이상 3만원 미만 : 1  
3만원 이상 5만원 미만 : 2  
5만원 이상 7만원 미만 : 3

```python
data.head(2)
```

```python
def clean(df):
    return df.replace('구매 의사 없음', '0').replace('1만원 이상 3만원 미만', '1').replace('3만원 이상 5만원 미만', '2').replace('5만원 이상 7만원 미만', '3')

data['선호 가격대(clean)'] = data['선호 가격대'].apply(clean)
data.head(3)
```

```python
data['선호가격대(clean)'] = data['선호 가격대(clean)'].astype(int)
```

```python
sns.set_style('whitegrid')
plt.rc('font', family='Malgun Gothic')
sns.scatterplot(x=data.index,
                y='선호 가격대(clean)',
                data=data, hue='성별', palette=palette)
plt.grid(color='lightgrey', alpha=0.3)
```

```python
real = pd.DataFrame(data=[0, 0, 50, 33, 0],
            index=['뚱카롱(59,000원)', '책&영화 소셜클럽(199,000원)', '포토샵(30,000원)',
                   '캐릭터방향제&캔들제작(25,000원)', '북한산&도시락(31,500원)'],
            columns=['실제 소비자가 될 확률(실제 가격<선호 가격대)'])
real
```

```python
sns.set_style('whitegrid')
plt.rc('font', family='Malgun Gothic')

real.plot(kind='line', figsize=(12,8), color='r')
plt.title('선택한 프로그램의 선호 가격대와 실제 가격의 차이를 통해 실제 소비자가 될 확률')
plt.xlabel('선호 프로그램')
plt.ylabel('퍼센트(%)')
plt.grid(color='lightgrey', alpha=0.3)
plt.tight_layout()
```
